### Simple Linear Regression
Before moving forward, it is interesting to look at the linear regressions of one feature and the price. 

```python
import scipy.stats as stats
results = [['ind_var', 'r_squared', 'intercept', 'slope', 'p-value', 'normality (JB)' ]]
   
    
for i, val in enumerate(['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
                        'waterfront','view','condition','grade','sqft_above',
                        'yr_built','lat','long','sqft_living15','sqft_lot15']):
    print ("King County Data Set - Regression Analysis for price~" + val)
    
    f = 'price~' + val
    model = ols(formula=f, data=df).fit()
    
    new_val = pd.DataFrame({val : [df[val].min(),df[val].max()]})
    predictions = model.predict(new_val)
    df.plot(kind='scatter', x = val, y = 'price',figsize=(4,4))
    plt.plot(new_val,predictions,color='red')
    plt.show()

    results.append([val, model.rsquared, model.params[0], model.params[1], model.pvalues[1], sms.jarque_bera(model.resid)[0]])
    
```

```python
pd.DataFrame(results)
```

From the graph and and the dataframe results, price has the highest correlation to square foot of living space at 0.50, followed by grade, sqft_above, sqft_living15, grade and bathrooms.

The stepwise_selection code is not my own, it was copied from the lab.

```python
import statsmodels.api as sm

def stepwise_selection(X, y, 
                       initial_list=[], 
                       threshold_in=0.01, 
                       threshold_out = 0.05, 
                       verbose=True):
    """ Perform a forward-backward feature selection 
    based on p-value from statsmodels.api.OLS
    Arguments:
        X - pandas.DataFrame with candidate features
        y - list-like with the target
        initial_list - list of features to start with (column names of X)
        threshold_in - include a feature if its p-value < threshold_in
        threshold_out - exclude a feature if its p-value > threshold_out
        verbose - whether to print the sequence of inclusions and exclusions
    Returns: list of selected features 
    Always set threshold_in < threshold_out to avoid infinite looping.
    See https://en.wikipedia.org/wiki/Stepwise_regression for the details
    """
    included = list(initial_list)
    while True:
        changed=False
        # forward step
        excluded = list(set(X.columns)-set(included))
        new_pval = pd.Series(index=excluded)
        for new_column in excluded:
            model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included+[new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min()
        if best_pval < threshold_in:
            best_feature = new_pval.idxmin()
            included.append(best_feature)
            changed=True
            if verbose:
                print('Add  {:30} with p-value {:.6}'.format(best_feature, best_pval))

        # backward step
        model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included]))).fit()
        # use all coefs except intercept
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max() # null if pvalues is empty
        if worst_pval > threshold_out:
            changed=True
            worst_feature = pvalues.argmax()
            included.remove(worst_feature)
            if verbose:
                print('Drop {:30} with p-value {:.6}'.format(worst_feature, worst_pval))
        if not changed:
            break
    return included
```

# 7. Summary 

This project used a data set which provided 18 key features for houses sold between May of 2104 and May of 2015 in Kings Country, Washington. The goal of the project was to build a multivariate linear regession model that could accurately predict the sale price of houses based on top key features.

The first step of the model was to import, evaluate and clean the data. Following is a summary of what transpired:

- The datatype for date was corrected, the start and stop date of the data set were discovered
- The dataype for zipcode was changed from integer(int) to string (str). My goal was to group the 70 zipcodes into "bins" so that the zipcode data could be used for the regression as dummy variables because it is categorical.
- Lat and long values represent a zipcode. I created a heat map to plot each house according to latitude and longitude to represent where the sold houses were located on a map.
- 2% of sqft_basement values were null. I choose not to delete the data, but did not use this feature because of redundancy (the sum of basement and above is equal to living).
- View and condition were not used because they lacked the integrity of the grade feature, which was a value officially assigned by Kings County.
- Less than 1% of the houses had a waterfront view. The null values were converted to 0, indicating no waterview for a house.
- The datatype for yr_renovated was changed to int to match that of yr_built. About 20% of the data values were missing, but I choose not to delete it. My desire was to covert the null values in the ratio of no renovation/renovation, but did not have the coding skills.
- The id feature was checked for duplicates and it was discovered that the duplicates where the same houses that had sold and then been re-sold again at a later date. Id was not used in the regression.
- One anomoly was found in bedrooms, a value of 33 bedrooms was changed to 3.  It was probably a missed keystroke.
- Price is the target feature. It was checked for normality. There were outliers of the price data which occured in the \$1,000,000+ range.

Following cleaning, the data was checked for multicollinearity (sqft_living was problematic) and four linear regression iterations were performed on four different models.

1. Sqft_living was excluded from the list of features
2. Sqft_living was added to the list of features
3. Sqft_above was subsituted for sqft_living in the list of features
4. Lat and long were added to the model, sqft_living was included in the features.

The fourth model faired the best with an R-squared of 0.55 using Statsmodel. The adjusted R-squared value was very close to the R-squared indicating that all the features contribute to explaining the variation in price. The values were similar using sci-kit learn's model. The result of using the stepwise function, the top 5 features of model 4 are:

1. bathroom
2. yr_built
3. sqft_living
4. longitude
5. floors

The p-values are (much) less than 0.05, which means there is evidence of a linear relationship between the feature and the price.


# 8. Conclusion

The fourth iteration created the highest R-squared value of 55% with the Statsmodel and 52% with sci-kit learn.  The 5 highest ranking features are: bathroom, yr_built, sqft_living, longitude and floors. 

When the correlation matrix was applied to the features before any linear regressions were performed, it was found that sqft_living correlated significantly with bathrooms, bedrooms, sqft_above and sqft_living15. Therefore, in this model it makes sense that when the sqft_living coefficient increases (1.503e+06) so will the bathroom and floor coefficients (8.37e+04 and 7.40e+04, respectively). Presumably when a house gets bigger in living space, more rooms are added. As living space increases, the longitude decreases (coefficient of -2.522e+05)  which means the house moves closer west to the coast line and water.  As seen earlier in the project, house prices are higher when there is a waterview.  Finally, as house prices increase the coefficient for yr_built decreases (-2.49e+05) meaning the house is newer and has higher grade features.  

The results of this model has room for improvement: only 55% of the variation in the dependent variable (price) is explained by the independent variable (features) in the model. The model is underfitted. Besides error in feature selection, the low R-squared value can be explained by insufficient or incorrect use of transformations. Another reason the R-squared value may be low is that it can be hard to predict human behavior. Finally, there may be other features - those that are not quantifiable - that play into house sales. 

From a home buyers perspective, it makes sense that you would choose to pay more for a house with more bathrooms, bedrooms and floors.  From an investors perspective, house prices can be predicted by looking at the number of bathrooms in a house, size of living space, the longitude (closer to the water) and age of the houses built.
